```python
import mne

# Replace 'your_file.edf' with the actual filename
raw = mne.io.read_raw_edf('dataset_change.edf')
```

```python
import numpy as np

# Replace 'your_file.edf' with the actual filenam

# Extract data and header information
data = raw.get_data()
header = ','.join(raw.info['ch_names'])

# Save data to CSV file
np.savetxt('data.csv', data.T, delimiter=',', header=header, fmt='%s')
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Display the DataFrame
df
```

```python
from matplotlib import pyplot as plt
df['EEG C3-A2'].plot(kind='hist', bins=20, title='EEG C3-A2')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
df['# EEG A1-A2'].plot(kind='hist', bins=20, title='# EEG A1-A2')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python


# Get information about channels
info = raw.info

# Loop through each channel and print relevant information
for ch_name in info['ch_names']:
    ch_info = info['chs'][info['ch_names'].index(ch_name)]
    print(f"Channel: {ch_name}")
    print(f"  Sampling frequency: {ch_info['sfreq']} Hz")
    print(f"  Physical minimum: {ch_info['range'][0]}")
    print(f"  Physical maximum: {ch_info['range'][1]}")
    print(f"  Digital minimum: {ch_info['cal'][0]}")
    print(f"  Digital maximum: {ch_info['cal'][1]}")
    print()
```
